```python
import pandas as pd

data = {
    "Producto": ["Laptop", "Celular", "Tablet", "Laptop", None, "Celular"],
    "Precio": [1500, 800, 600, 1400, 1000, None],
    "Cantidad": [10, 20, 15, None, 5, 30]
}
df = pd.DataFrame(data)
df
```

```python
df[['Precio','Cantidad']] = df[['Precio','Cantidad']].fillna(0)
df
```

```python
df['Total'] = df['Precio'] * df['Cantidad']
df
```

```python
df.sort_values(by='Total', ascending=False,inplace=True)
df
```
